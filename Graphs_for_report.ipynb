# Graphs of phase and time

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#The command above tells matplotlib to put the plots in a notebook
```

```python
delta_t_beta, phase_beta, vel_shift_beta = np.loadtxt("beta_cma_time_phase_vel_shift_Si3.csv", unpack=True, delimiter=',') 
#this is the data for beta CMa

vel_shift_err_beta = np.loadtxt("beta_photospheric_errors.txt", unpack=True)

delta_t_xi1, phase_xi1, vel_shift_xi1 = np.loadtxt("xi1_cma_time_phase_vel_shift_Si3.csv", unpack=True, delimiter=',')
#this is the data for xi1 CMa

vel_shift_err_xi1 = np.loadtxt("xi1_photospheric_errors.txt", unpack=True)

print(vel_shift_err_beta[0])
print(vel_shift_err_beta[7])
print(vel_shift_err_beta[14])
print(vel_shift_err_beta[21])
print(vel_shift_err_beta[28])
print(vel_shift_err_beta[35])
print(vel_shift_err_beta[41])
```
